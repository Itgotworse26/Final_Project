```python
import pandas as pd
import matplotlib.pyplot as plot
```

```python
# loading csv file
car_df=pd.read_csv("Resources/Car details v3.csv")
# car_df.head(5)
print(car_df.count)
```

```python
# dropna for each row
car_dropna_df= car_df.dropna(how="any")
# car_dropna_df.head(20)
print(car_dropna_df.count)
```

```python
car_dropna_df.head(20)
```

```python
# Seperate name and split it between manufacturer and model
car_dropna_df[['manufacturer','type']] =  car_dropna_df["name"].str.split(" ", n=1, expand=True)
car_dropna_df.head()
```

```python
# Select column
car_clean_df=car_dropna_df[['brand',"year","selling_price","fuel","seller_type","mileage"]]
car_clean_df
```

```python
# Clean fuel types
car_clean_df['fuel'] = car_clean_df['fuel'].replace(['CNG', 'LPG'], 'Other')
```

```python
car_clean_df.dtypes
```

```python
car_over19=car_clean_df[car_clean_df["year"]>2017]

print(car_over19.count)
```

```python
# save clean data 
car_over19.to_csv("Resources/car_clean.csv")
```
